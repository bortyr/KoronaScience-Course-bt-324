```python
def rectangle_area(a, b):
  if a<0 or b<0:
    raise ValueError("if either number was negative")
  return a*b
```

*Przykładowe operacje na danych o koronawirusie*

```python
url = f"https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv"
df = pd.read_csv(url, error_bad_lines=False)
```

```python
df
```

```python
#liczba przypadków - miejsca z największą liczbą przypadków
def top(data):
  return df[["Province/State", "Country/Region", data]].sort_values(by=data).tail(10)
```

```python
data14="3/15/20"
```

```python
top(data14)
```

```python
#kolorowe kredki dla DataFrame
df.loc[df["Country/Region"]=="China",:].sort_values(by=data14).tail(10).head(9).style.background_gradient(cmap='magma')
```

```python
#brak przypadkow do danego dnia (włacznie)
def brakwirusa(data):
  return df.loc[df[data]==0].index
```

```python
brakwirusa(data14)
```

*Mapa przypadków na świecie*

```python
import plotly.express as px
```

```python
formated_gdf = df.groupby(["Country/Region"]).max()
```

```python
formated_gdf =  formated_gdf.reset_index()
```

```python
formated_gdf['size'] = formated_gdf[data14].pow(0.3)
```

```python
formated_gdf
```

```python
formated_gdf = df.groupby
```

```python
fig = px.scatter_geo(formated_gdf, locations="Country/Region", locationmode='country names', 
                     color="3/14/20", size="size", hover_name="Country/Region", 
                     range_color= [0, max(formated_gdf['3/14/20'])+2], 
                     projection="natural earth", 
                     title="COVID-19 na swiecie" 
                     )

fig.show()
```
